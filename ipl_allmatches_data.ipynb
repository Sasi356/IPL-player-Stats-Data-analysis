# Data preprocessing

```python
dataset = read_csv('all_matches.csv')
data = dataset.drop(['match_id','other_player_dismissed','other_wicket_type','penalty','byes','start_date'],axis =1)
data['wides'] = data['wides'].fillna(0)
data['noballs'] = data['noballs'].fillna(0)
data['legbyes'] = data['legbyes'].fillna(0)
data['player_dismissed'] = data['player_dismissed'].fillna('not out')
data['wicket_type'] = data['wicket_type'].fillna('not out')
data
```

# ---------------------------------------Batting stats---------------------------------------------

# Total Runs

```python
batsmen = data['striker'].unique()
batsmen
```

```python
#Before sorting
runs_list = {}
for player in batsmen:
    d = data.loc[data['striker']==player]
    runs_list[player] = sum(d['runs_off_bat'])
runs_list
```

```python
#Let us sort the table from highest to lowest
s_runs_list= list(runs_list.values())
s_runs_list.sort(reverse = True)
#print(s_runs_list)
#After sorting
total_runs = {}
for score in s_runs_list:
    for player in batsmen:
        if runs_list[player]==score:
            total_runs[player] = score
total_runs
```

# Strike rate

```python
strike_rate_list = {}
batsmen_=[]
for player in batsmen:
    d = data.loc[data['striker']==player]
    if len(d)>=100:
        batsmen_.append(player)
        strike_rate_list[player] = sum((d['runs_off_bat']/len(d)*100))

print(len(strike_rate_list))
strike_rate_list
```

```python
s_sr_list = list(strike_rate_list.values())
s_sr_list.sort(reverse = True)
s_sr_list
```

```python
strike_rate = {}
for rate in s_sr_list:
    for player in batsmen_:
        if strike_rate_list[player]== rate:
            strike_rate[player] = rate
strike_rate
```
